```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer, PorterStemmer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Label'] = le.fit_transform(data['Label'])
```

```python
data.head()
```

```python
y = data['Label']
data.drop(['Label'], inplace=True, axis=1)
```

```python
ps = PorterStemmer()
wnl = WordNetLemmatizer()
```

```python
import re
corpus=[]
for element in data['Statement']:
    final_txt = re.sub('[^a-zA-Z]', ' ', element)
    final_txt = final_txt.lower()
    final_txt = final_txt.split()
    final_txt = [ps.stem(word) for word in final_txt if not word in stopwords.words('english')]
    final_txt = ' '.join(final_txt)
    corpus.append(final_txt)
print(corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(corpus)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
model = MultinomialNB()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model.fit(X_train, y_train)
predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, predict)
print(score)
```

```python
import pickle 
file = open('FakeNewsModel.pkl','wb')
pickle.dump(model,file)
```
